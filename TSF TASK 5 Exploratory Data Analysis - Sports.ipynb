```python
matches_2017 = data[data["season"]==2017]
```

```python
matches_2019 = data[data["season"]==2019]
```

```python
#Now we are visualizing the player who has won most player of the match awards (season: 2008-2017)

plt.subplots(figsize=(10,10))
data["player_of_match"].value_counts().head(7).plot.bar()
plt.xlabel("Player")
plt.ylabel("Count")
plt.title("most player of the match award")
plt.show()
```

Chris Gayle has won most player of the match awards. 

```python
#Team which have won most number of matches

plt.subplots(figsize=(10,8))
data['winner'].value_counts().plot.bar(width=0.8)
plt.xlabel("TEAMS")
plt.ylabel("Count")
plt.title("Team winning matches most number of times")
plt.show()
```

MUMBAI INDIANS team have won most number of matches. 

```python
#Choice of Toss in different Seasons 

plt.subplots(figsize=(10,8))
sns.countplot(x="season",hue="toss_decision",data=data)
plt.title("Choice of Toss")
plt.legend(title='Toss Decision')
plt.show()
```

```python
#THE CHOICE TO SELECT WHETHER TO TAKE BAT OR FIELD DIFFER IN DIFFERENT SEASONS

#How many matches the team has won if they had won the toss and if they have lost the match then to whom they lost it 
# (Season: 2017)

b=list(matches_2017["toss_winner"].unique())
fig,axes1=plt.subplots(figsize=(20,14))
sns.countplot(x="toss_winner",data=matches_2017,hue="winner" ,ax=axes1)

axes1.set_xticklabels(b,rotation=90)
axes1.set_xlabel("")
```

```python
#THE CHOICE TO SELECT WHETHER TO TAKE BAT OR FIELD DIFFER IN DIFFERENT SEASONS

#How many matches the team has won if they had won the toss and if they have lost the match then to whom they lost it 
# (Season: 2019)



b=list(matches_2019["toss_winner"].unique())
fig,axes1=plt.subplots(figsize=(20,14))
sns.countplot(x="toss_winner",data=matches_2019,hue="winner" ,ax=axes1)

axes1.set_xticklabels(b,rotation=90)
axes1.set_xlabel("")
```

```python
#Now we analyse that whether Teams win the match or not if they had won the Toss

df=data[data['toss_winner']==data['winner']]
slices=[len(df),(577-len(df))]
labels=['won','loss']
plt.pie(slices,labels=labels,startangle=90,autopct='%1.1f%%',colors=['g','r'],shadow=True, explode=(0,0.1))
plt.legend()
plt.show()
```
